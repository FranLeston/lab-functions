## 2. Now write a function that returns the largest element on a list

```python
def greatest(arr):
    is_max = arr[0]
    for num in arr:
        if num > is_max:
            is_max = num
    return is_max

greatest([3,4,5,6,65])        
```

```python
# This will test your function 
test_greatest(greatest)
```

## 5. Now combine those two ideas and write a function that receives a list and either "+" or "*" and outputs acordingly

```python
def oper_all(arr, oper):
    if oper == "+":
        sum = 0
        for num in arr:
            sum += num
        return sum
    if oper == "*":
        product = 1
        for num in arr:
            product *= num
        return product
        
oper_all([2,3,4], "*")
```

## 6. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
# This will test your function 
test_factorial(factorial)
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. ü§î`

```python
def unique(arr):
    my_list = []
    for el in arr:
        if el not in my_list:
            my_list.append(el)
    return my_list
```

```python
# This will test your function 
test_unique(unique)
```

## 9. Write a function that calculates the standard deviation of a list.
`NOTE: Do not use any libraries or already built functions. üòâ`

```python
def st_dev(arr):
    media=sum(arr)/len(arr)
    var = sum((n-media)**2 for n in arr)
    std=(var/(len(arr)-1))**(1/2)
    return std

st_dev([1,2,3,4,5,6])
```

```python
# This will test your function 
test_stdev(st_dev)
```

## 10. Write a function to check if a string is a pangram, i.e.: if it contains all the letters of the alphabet at least once. Mind that the strings may contain characters that are not letters.

```python
def pangram(string):
    my_alphabet = "abcdefghijklmnopqrstuvwxyz"
    for letter in my_alphabet:
        if letter not in string.lower():
            return False
    return True
pangram("The quick brown fox jumps over the lazy dog 5")
```

```python
# This will test your function 
test_pangram(pangram)
```

## 11. Write a function that receives a string of comma separated words and returns a string of comma separated words sorted alphabetically.

`NOTE: You may use sorted but not split and definitely no join! ü§™`

```python
def sort_alpha(string):
    split_value = []
    tmp = ''
    for char in string:
        if char == ',':
            split_value.append(tmp)
            tmp = ''
        else:
            tmp += char
    if tmp:
        split_value.append(tmp)
    
    sorted_list = sorted(split_value, key = lambda x: x[0])

    result = ''
    last = len(sorted_list) - 1
    for pos, elem in enumerate(sorted_list):
        result += str(elem)
        if pos != last:
            result += ', '
    return result

print(sort_alpha("hello,my,name,is,fran,5,a,b,c,Z"))

#I get a final string sorted alphabetically but cannot pass the test
#ALSO..a Capital 'Z' will go before a lowecase 'a'.  Would love to see this one explained! 
```

```python
# This will test your function 
test_alpha(sort_alpha)
```

## 12. Write a function to check if a given password is strong (at least 8 characters, at least one lower case, at least one upper case, at least one number and at least one special character). It should output True if strong and False if not.
`Valid special characters: # @ ! $ % & ( ) ^ * [ ] { }`

```python
from string import printable

def check_pass(string):
    
    special_chars = ['#', '@', '!', '$', '%', '&', '(', ')', '^', '*', '[', ']', '{', '}']
    lower_case_count = 0
    upper_case_count = 0
    number_count = 0
    special_count = 0
    
    for char in string:
        if char.islower():
            lower_case_count += 1
        if char.isupper():
            upper_case_count += 1
        if char.isnumeric():
            number_count += 1
        if char in special_chars:
            special_count += 1
    
    if len(string) >= 8 and lower_case_count > 0 and upper_case_count > 0 and number_count > 0 and special_count > 0:
        return True
    else:
        return False
print(check_pass("Pepitos3%"))
```

```python
# This will test your function 
test_pass(check_pass)
```
